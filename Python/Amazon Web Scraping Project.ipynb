```python
# Importing Libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
import pandas as pd
```

```python
# Connecting to website & pulling data

url = 'https://www.amazon.co.uk/Things-Data-Scientist-Analyst-T-Shirt/dp/B0DCNNFKQZ?customId=B07536XX75&customizationToken=MC_Assembly_1%23B07536XX75&th=1&psc=1'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:151.0) Gecko/20100101 Firefox/151.0"}

page = requests.get(url, headers=headers)          

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find(id="productTitle").get_text()

price = soup2.find('span', class_='aok-offscreen').get_text(strip=True)

print(title)
print(price)









```

```python
# Cleaning up data so it can be imported into an Excel file

price = price.strip()[1:]
title = title.strip()

print(title)
print(price)
```

```python
# Creating a timestamp as to track data collection time

import datetime 

today = datetime.date.today()
print(today)
```

```python
# Creating the CSV file and writing the headers and data into the file
import csv

header = ['Title', 'Price', 'Date']
data = [title, price, today]

with open('AmazonWebScrapingDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
# Combining above code into a function

def check_price():
    url = 'https://www.amazon.co.uk/Things-Data-Scientist-Analyst-T-Shirt/dp/B0DCNNFKQZ?customId=B07536XX75&customizationToken=MC_Assembly_1%23B07536XX75&th=1&psc=1'

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:151.0) Gecko/20100101 Firefox/151.0"}

    page = requests.get(url, headers=headers)          

    soup1 = BeautifulSoup(page.content, "html.parser")

    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

    title = soup2.find(id="productTitle").get_text()

    price = soup2.find('span', class_='aok-offscreen').get_text(strip=True)

    price = price.strip()[1:]
    title = title.strip()

    import datetime 

    today = datetime.date.today()

    import csv

    header = ['Title', 'Price', 'Date']
    data = [title, price, today]

    with open('AmazonWebScrapingDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)

    numeric_price = float(price) 
    
    if numeric_price < 16.00:  
        send_mail()
```

```python
# Runs check_price function every 24 hours and inputs data into the CSV file

while(True):
    check_price()
    time.sleep(86400)
```
